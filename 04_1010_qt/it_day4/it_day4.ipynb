## Windowを作る

第一歩として，ウィンドウタイトルのみで中身が何もないウィンドウを作成します．

1. [QMainWindow](http://pyqt.sourceforge.net/Docs/PyQt4/qmainwindow.html)クラスのインスタンス`w`を生成します．
2. メインウィンドウの大きさやウィンドウタイトルを変更するためにメソッド[resize()](http://pyqt.sourceforge.net/Docs/PyQt4/qwidget.html#resize-2), [setWindowTitle()](http://pyqt.sourceforge.net/Docs/PyQt4/qwidget.html#setWindowTitle)を呼び出します．<br>
![image](img/day5_window.png)
3. ウィンドウを表示します: [show()](http://pyqt.sourceforge.net/Docs/PyQt4/qwidget.html#show)

フリーズしたり，動作が不安定になったら，jupyter notebookのメニュー [Kernel]-[Restart]でインタプリタを再起動すれば復帰するので，怖がらずにtry&errorをしてください．

````python
import sys
from PyQt5.QtWidgets import *

def main():
    app = QApplication(sys.argv)
    w = QMainWindow()       # 1. MainWindow()クラスのインスタンスを生成し変数wに代入
    w.resize(250, 150)            # 2. ウィンドウwを250x150ピクセルにリサイズ
    w.setWindowTitle('QtSample')  #   ウィンドウwのタイトルを設定
    w.show()                      #   ウィンドウwを表示（実際は，app.exec_()を呼ぶまで表示されない）
    app.exec_()

if __name__ == '__main__': 
    main()
````

* ``app.exec_()``を呼び出すとPyQtライブラリ内のループに入り，メインウィンドウ``app``が消滅するまで制御が``main()``に戻ってきません．CUI (Commandline User Interface)との最も大きな違いです．

````python
import sys
from PyQt5.QtWidgets import *

def main():
    app = QApplication(sys.argv)
    w = QMainWindow()       # 1. MainWindow()クラスのインスタンスを生成し変数wに代入
    # サイズを500x300にした
    w.resize(500, 300)            # 2. ウィンドウwを500x300ピクセルにリサイズ
    # タイトルを日本語にした
    w.setWindowTitle('日本語が使えるか試してみるとよい')  #   ウィンドウwのタイトルを設定
    w.show()                      #   ウィンドウwを表示（実際は，app.exec_()を呼ぶまで表示されない）
    # オプション
    w2 = QWidget()
    w2.resize(300, 200)
    w2.setWindowTitle('sub window')
    w2.show()
    app.exec_()

if __name__ == '__main__': 
    main()
````

## ステータスバーに文字などを表示する

[QMainWindow](http://pyqt.sourceforge.net/Docs/PyQt4/qmainwindow.html "QMainWindow")クラスの[statusBar()](http://pyqt.sourceforge.net/Docs/PyQt4/qmainwindow.html#statusBar)メソッドで[QStatusBar](http://pyqt.sourceforge.net/Docs/PyQt4/qstatusbar.html)オブジェクトを取得し，[ShowMessage()](http://pyqt.sourceforge.net/Docs/PyQt4/qstatusbar.html#showMessage)メソッドを呼び出すことで，ステータスバーに文字列を表示できます．

![statusbar](img/day5_statusbar.png)

````python
import sys
from PyQt5.QtWidgets import *
import datetime

def time_draw(w):
    """ 現在時刻をステータスバーに表示 """
    d = datetime.datetime.today()
    daystr=d.strftime("%Y-%m-%d %H:%M:%S")
    sb = w.statusBar()
    sb.showMessage(daystr)   # ステータスバーにメッセージを表示
    
def main():
    app = QApplication(sys.argv)
    w = QMainWindow()             # MainWindow()クラスのインスタンスを生成し変数wに代入
    w.resize(250, 150)            # ウィンドウwを250x150ピクセルにリサイズ
    w.setWindowTitle('QtSample')  # ウィンドウwのタイトルを設定
    w.show()                      # ウィンドウwを表示（実際は，app.exec_()を呼ぶまで表示されない）

    time_draw(w)   # <==

    app.exec_()

if __name__ == '__main__':
    main()
````

### クラスを使った実装
`MyWindow`クラスを`QMainWindow`の子クラス（派生クラス）として定義します．
`QMainWindow`クラスの属性とメソッドを継承し，差分は`initUI()`メソッドに定義しています．
* ウィンドウサイズを250, 150に変更
* ウィンドウタイトルを'MyWindow'に設定
* ステータスバーに現在時刻を表示

````python
import sys
from PyQt5.QtWidgets import *
import datetime

class MyWindow(QMainWindow):
    
    # 適当な文字列を表示するように変更した
    def __init__(self, daystr: str=datetime.datetime.today().strftime("%Y-%m-%d %H:%M:%S")):
        """ インスタンスが生成されたときに呼び出されるメソッド """
        self.daystr = daystr
        super(MyWindow, self).__init__()
        self.initUI()
    
    def initUI(self):
        """ UIの初期化 """
        self.resize(250, 150)           # 250x150ピクセルにリサイズ
        self.setWindowTitle('MyWindow') # タイトルを設定
        
        self.time_draw()                # 現在時刻をステータスバーに表示
        
        self.show()

    def time_draw(self):
        """ 現在時刻をステータスバーに表示 """
        self.statusBar().showMessage(self.daystr)   # ステータスバーを取得しsendMessage()メソッドを呼び出す

def main():
    app = QApplication(sys.argv)
    w = MyWindow(daystr='10月31日はハロウィーンです！')
    app.exec_()

if __name__ == '__main__':
    main()
````

## メニューバーを作る

以下の手順でメニューバーとメニュー項目[File]-[Exit]を作成します．変更したのは`initUI()`メソッドの9行だけです．
1. ニュー項目[File]-[Exit]が選択されたときのアクションを定義するために[QAction](http://pyqt.sourceforge.net/Docs/PyQt4/qaction.html)クラスのインスタンス`exitAction`を生成します．
2. メニュー項目が選択されたときに，ウィンドウを閉じる(`self.close`メソッド)を呼び出すように設定します．
3. メインウィンドウの[QMenuBar](http://pyqt.sourceforge.net/Docs/PyQt4/qmenubar.html)オブジェクトを取得します: `menubar = self.menuBar()`
4. メニューバーに[File]メニューを追加し，そのアクションとして`exitAction`を登録する．

![menubar](img/day5_menu.png)


````python
import sys
from PyQt5.QtWidgets import *
import datetime

class MyWindow(QMainWindow):
    
    def __init__(self):
        """ インスタンスが生成されたときに呼び出されるメソッド """
        super(MyWindow, self).__init__()
        self.initUI()
  
    def initUI(self):
        """ UIの初期化 """
        """↓ ここから """
        # (1) ニュー項目[File]-[Exit]が選択されたときのアクションを生成
        exitAction = QAction('&Exit', self)
        # (2) メニュー項目が選択されたときの処理として，qApp.quitを設定
        #exitAction.triggered.connect(qApp.quit)
        exitAction.triggered.connect(self.close)
        
        # (3) メインウィンドウのメニューバーオブジェクトを取得
        menubar = self.menuBar()
        # macの場合はこれを書かないとちゃんと動作しないよ
        from platform import system
        if system() == 'Darwin':
            menubar.setNativeMenuBar(False)
            del system
        # (4) メニューバーに[File]メニューを追加し，そのアクションとしてexitActionを登録
        fileMenu = menubar.addMenu('&File')
        fileMenu.addAction(exitAction)
        """ ↑ここまで"""
        
        self.time_draw()
        
        self.resize(250, 150)           # 250x150ピクセルにリサイズ
        self.setWindowTitle('MyWindow') # タイトルを設定
        self.show()
        
    def time_draw(self):
        """ 現在時刻をステータスバーに表示 """
        d = datetime.datetime.today()
        daystr=d.strftime("%Y-%m-%d %H:%M:%S")
        self.statusBar().showMessage(daystr)   # ステータスバーを取得しsendMessage()メソッドを呼び出す
        
def main():
    app = QApplication(sys.argv)
    w = MyWindow()
    app.exec_()

if __name__ == '__main__':
    main()
````

## ボタンを表示

以下の手順でボタンを作成し，ウィンドウにレイアウトします．

1. ボタンの数だけ，`QPushButton`クラスのインスタンスを作成する
2. 横方向にボタンを並べたい場合は`QHBoxLayouot`クラスのインスタンスを作成し，ボタンを追加する
3. 縦方向にボタンを並べたい場合は`QVBoxLayout`を使用する．
4. ボタン`button`が押されたときのアクションを`button.clicked().connect(action)`で登録する．

![画面](img/day5_buttons.png)

````python
import sys
from PyQt5.QtWidgets import *
import datetime

class MyWindow(QMainWindow):
    
    def __init__(self):
        """ インスタンスが生成されたときに呼び出されるメソッド """
        super(MyWindow, self).__init__()
        self.initUI()
    
    def on_click1(self):
        print("clicked!")
        
    def on_click2(self, text):
        print(text)
    
    def initUI(self):
        """ UIの初期化 """
        self.resize(250, 150)           # 250x150ピクセルにリサイズ
        self.setWindowTitle('MyWindow') # タイトルを設定
        
        # ↓ここから
        b1 = QPushButton('上')
        b2 = QPushButton('左')
        b3 = QPushButton('右')
        b4 = QPushButton('下')
        
        b1.clicked.connect(self.on_click1)                # 普通は引数なしで関数を呼び出す
        b2.clicked.connect(lambda: self.on_click2("左"))  # 引数を与えたい場合はlambda式を使う方法がある
        b3.clicked.connect(lambda: self.on_click2("右"))
        b4.clicked.connect(lambda: self.on_click2(b4.text()))

        # この部分に変更したコードを記述する
        hbox = QHBoxLayout()
        hbox.addStretch(1)      # <-- ★1
        # この値を大きくすると、ボタン「左、右」が右側へ移動し、逆に小さくすると左側へ移動した場所に表示されるようになる。
        hbox.addWidget(b1)
        hbox.addWidget(b2)
        hbox.addWidget(b3)
        hbox.addStretch(1)
        
        vbox = QVBoxLayout()
        vbox.addLayout(hbox)
        vbox.addWidget(b4)
        
        container = QWidget()
        container.setLayout(vbox)

        self.setCentralWidget(container)
        
        # ↑ここまで
        self.show()

def main():
    app = QApplication(sys.argv)
    w = MyWindow()
    app.exec_()

if __name__ == '__main__':
    main()
````

> 2. ボタン`b1`をボタン`b2`の左側に表示するように上のプログラムを変更してください．つまり，上ボタンが左ボタンの左に表示される．

## メッセージボックスを表示
[QMessageBox](http://pyqt.sourceforge.net/Docs/PyQt4/qmessagebox.html)クラスのメソッドを呼び出すことで各種メッセージボックスを表示できる．**以下を実行すると，メッセージボックスがウィンドウの裏に隠れてしまうことがあるので注意 **
1. [QMessageBox.question](http://pyqt.sourceforge.net/Docs/PyQt4/qmessagebox.html#question-2)![question](img/day5_messagebox_question.png)
2. [QMessageBox.warning](http://pyqt.sourceforge.net/Docs/PyQt4/qmessagebox.html#warning-2) ![warning](img/day5_messagebox_warning.png)
3. [QMessageBox.information](http://pyqt.sourceforge.net/Docs/PyQt4/qmessagebox.html#information-2) ![information](img/day5_messagebox_information.png)
4. [QMessageBox.critical](http://pyqt.sourceforge.net/Docs/PyQt4/qmessagebox.html#critical): 他にも色々あります．
5. [QMessageBox.about](http://pyqt.sourceforge.net/Docs/PyQt4/qmessagebox.html#about)

````python
import sys
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *

class MyWindow(QMainWindow):
    
    def __init__(self):
        """ インスタンスが生成されたときに呼び出されるメソッド """
        super(MyWindow, self).__init__()
        self.initUI()
    
    def initUI(self):
        """ UIの初期化 """        
        self.resize(250, 150)           # 250x150ピクセルにリサイズ
        self.setWindowTitle('MyWindow') # タイトルを設定 
        self.show()

        """ ↓ここから """
        # message box
        result = QMessageBox.question(self, 'Message', 
                    "PyQtに慣れましたか?", QMessageBox.Yes | QMessageBox.No, QMessageBox.No) 
        if result == QMessageBox.Yes:
            print ('Selected Yes.')
        else:
            print ('Selected No.')

        # Warning Message box
        QMessageBox.warning(self, "警告", "Warning! 何かおかしいようです")

        # Information Message box
        QMessageBox.information(self, "情報", "Please contact at hoge@gmail.com")
        """ ↑ここまで """

def main():
    app = QApplication(sys.argv)
    w = MyWindow()
    app.exec_()

if __name__ == '__main__':
    main()
````

2. [QGraphicsView](http://pyqt.sourceforge.net/Docs/PyQt4/qgraphicsview.html)を
カスタマイズするために新しく**MyGraphicsViewクラス**を定義します．
MyGraphicsViewクラスの**mouseMoveEvent()メソッド**は，
マウスボタンを押しながらマウスが移動したときに発生するイベントの処理を定義しています．
  1. `__init__()`メソッド：ウィジェットの初期化．後で使うペン`QPen`, ブラシ`QBrush`のインスタンスを作成しておきます．
  2. `mouseMoveEvent()`メソッド：ボタンを押した状態でマウスが移動すると呼び出される．
  マウスの位置`pos`を取得して，その位置に円を描画します．




````python
import sys
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtCore import *

class MyWindow(QMainWindow):
    
    def __init__(self):
        """ インスタンスが生成されたときに呼び出されるメソッド """
        super(MyWindow, self).__init__()
        self.initUI()
    
    def initUI(self):
        """ UIの初期化 """        
        self.resize(500, 500)               # ウィンドウサイズを設定
        self.setWindowTitle('MyWindow')     # タイトルを設定 

        """ (1) QGraphicsView, QGraphicsSceneの作成 """ 
        self.view = MyGraphicsView()        # 図形描画表示のためのウィジェット
        self.scene = QGraphicsScene(self)   # 図形描画のためのウィジェット
        self.view.setScene(self.scene)      # self.viewとsceneを関連付ける
        self.view.setSceneRect(0, 0, 500, 500)  # シーン座標系の描画範囲
        self.setCentralWidget(self.view)    # QMainWindowの中央部に表示するウィジェットとして登録
        
        self.show()
        
class MyGraphicsView(QGraphicsView):
    def __init__(self):
        """ (2.A) ウィジェットの初期化 """
        super(MyGraphicsView, self).__init__()
        self.color_value = int(0xfff)              # 色決めのための値(0～0xfff)
        self.pen = QPen(QColor(0xff, 0xff, 0xff))  # ペンを作成
        self.brush = QBrush(QColor(0xff, 0xff, 0xff), Qt.SolidPattern)    #ブラシを作成
        
    def mouseMoveEvent(self, event):
        """ (2.B) ボタンを押した状態でマウスが動くと呼び出される """
        viewport_pos = event.pos()          # viewport座標系の座標値を取得
        pos = self.mapToScene(viewport_pos) # シーン座標系に変換
        x = pos.x()
        y = pos.y()
        # print "mousePressEvent: (", pos.x(), ",", pos.y(), ")"
        
        # 図形の表示色を設定
        cv = self.color_value = (self.color_value + 8) & 0xfff
        color = QColor((cv & 0xf00) >> 4, (cv & 0x0f0), (cv & 0x00f) << 4)
        self.pen.setColor(color)
        # self.brush.setColor(color)
        
        # 図形アイテムを作成しシーンに登録
        item = QGraphicsEllipseItem(x-10, y-10, 20, 20) # 楕円アイテムを作成
        item.setPen(self.pen)                           # ペンを設定
        item.setBrush(self.brush)                       # ブラシを設定
        self.scene().addItem(item)                      # Sceneに配置
            
def main():
    app = QApplication(sys.argv)
    w = MyWindow()
    app.exec_()

if __name__ == '__main__':
    main()
````

## スライダーを利用したGUIを作成
1. [QLabel](http://pyqt.sourceforge.net/Docs/PyQt4/qlabel.html)と[QSlider](http://pyqt.sourceforge.net/Docs/PyQt4/qslider.html)のインスタンス（ウィジェット）を生成する．"slider(%):"の文字列とスライダーが対応します．
2. スライダの値が変わったときに`on_draw()`メソッドを呼び出すよう設定する．
3. ウィジェットを水平方向に整列させる[QHBoxLayout](http://pyqt.sourceforge.net/Docs/PyQt4/qhboxlayout.html)クラスのインスタンス`hbox`を生成し，`slider_label`, `slider`ウィジェットを追加する．
4. 1行テキスト編集を行う[QLineEdit](http://pyqt.sourceforge.net/Docs/PyQt4/qlineedit.html)クラスのインスタンス（ウィジェット）を生成する．
5. ウィジェットを垂直方向に整列させる[QVBoxLayout](http://pyqt.sourceforge.net/Docs/PyQt4/qvboxlayout.html)のインスタンス`vbox`を生成し，`textbox`と`hbox`を追加する．
6. ウィンドウに`vbox`を設定する．

</table>


````python
import sys
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *

class MyWindow(QWidget):
    
    def __init__(self):
        """ インスタンスが生成されたときに呼び出されるメソッド """
        super(MyWindow, self).__init__()
        self.initUI()
    
    def initUI(self):
        """ UIの初期化 """        
        self.resize(250, 150)           # 250x150ピクセルにリサイズ
        self.setWindowTitle('MyWindow') # タイトルを設定 
       
        """ ↓ここから ****** """
        # (1) QLabelとQSliderのWidget（インスタンス）を生成
        slider_label = QLabel('Slider (%):') # ラベルを生成
        self.slider = QSlider(Qt.Horizontal) # 横向きのスライダを生成
        self.slider.setRange(0, 100)         # スライダの値の範囲を設定
        self.slider.setValue(20)             # 初期値を設定        
        # (2) sliderの値が変わったときにon_draw()メソッドを呼び出すよう設定
        self.slider.valueChanged.connect(self.on_draw) 
        # (3) QHBoxLayout(Widget1を水平方向に並べる)を生成し，slider_label, slider
        #     widgetsを追加
        hbox = QHBoxLayout()
        hbox.addWidget(slider_label)
        hbox.addWidget(self.slider)
        # (4) QLineEdit（1行テキスト編集)のwidget(インスタンス)を生成
        self.textbox = QLineEdit()        
        # (5) QVBoxLayout(Widetを垂直方向に並べる)を生成し，textboxとhboxを追加
        vbox = QVBoxLayout()
        vbox.addWidget(self.textbox)
        vbox.addLayout(hbox)        
        # (6) ウィンドウにvboxレイアウトを設定
        self.setLayout(vbox)
        """ ↑ここまで ****** """       
        self.show()
 
    """ ↓ここから ****** """
    def on_draw(self):
        """ スライダの値が変更されたときに呼び出されるメソッド """
        self.textbox.setText(str(self.slider.value()))
    """ ↑ここまで ****** """

def main():
    app = QApplication(sys.argv)
    w = MyWindow()
    app.exec_()

if __name__ == '__main__':
    main()
````
